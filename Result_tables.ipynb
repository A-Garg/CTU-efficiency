```python
diagnoses
```

```python
ICD_dictionary = {
    'A':'infectious',
    'C':'neoplasms',
    'D':'blood and immune',
    'E':'endocrine and metabolic',
    'F':'mentall, behavioural, neurodevelopmental',
    'G':'nervous system',
    'H':'eye',
    'I':'circulatory',
    'J':'respiratory',
    'K':'GI',
    'L':'skin/soft tissue',
    'M':'MSK/connective tissue',
    'N':'GU',
    'O':'pregnancy',
    'P':'perinatal',
    'Q':'congenital',
    'R':'abnormal clinical/labs',
    'S':'poisoning',
    'U':'special codes',
    'V':'external causes',
    'Z':'healthcare contact'
}
ICD_dictionary
```

```python
diagnoses['ICD category'] = diagnoses['First letter'].map(ICD_dictionary)
diagnoses
```

```python
group = diagnoses.groupby('Drip?')
group['ICD category'].value_counts(normalize=True)
```
